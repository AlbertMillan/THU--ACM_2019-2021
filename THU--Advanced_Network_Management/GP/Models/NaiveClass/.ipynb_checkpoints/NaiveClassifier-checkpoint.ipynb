```python
def format_df(df_func):
    
    if df_func.empty:
        df_func = pd.DataFrame(np.array([[0, 0], [1,0]]), columns=['label','freq'])
        df_func = df_func.set_index('label')
    
    elif df_func.shape[0] == 1:
        idx_c = 0

        idx_v = df_func.index[0]
        val_v = np.array(df_func)[0]

        if idx_v == 0:
            idx_c = 1

        df_func = pd.DataFrame(np.array([[idx_v, val_v], [idx_c,0]]), columns=['label','freq'])
        df_func = df_func.sort_values(by='label')
        df_func = df_func.set_index('label')

    else:
        vals = np.array(df_func)
        df_func = pd.DataFrame([[df_func.index[0], vals[0]], [df_func.index[1],vals[1]]], columns=['label','freq'])
        df_func = df_func.sort_values(by='label')
        df_func = df_func.set_index('label')

    return df_func
```

```python
input_dir = './../train/KPI/'
# up_limit = 0.5
# low_limit = 0.5

summary = pd.DataFrame(columns=['St', 'Anom', 'St_up', 'Anom_up', 'St_down', 'Anom_down'])


arr = [[None,0,0,0,0,0,0] for x in range(26)]
up_limit = 0.49
low_limit = 0.49

while up_limit < 3:
    up_limit += 0.01
    low_limit += 0.01
    up_limit = round(up_limit,2)
    low_limit = round(low_limit,2)
#     summary = pd.DataFrame(columns=['St', 'Anom', 'St_up', 'Anom_up', 'St_down', 'Anom_down'])
    
    i=0
    
    for fname in os.listdir(input_dir):
        df  = pd.read_csv(os.path.join(input_dir, fname), index_col='timestamp')

        # Check Labels
        x = df.label.groupby(df.label).count()
        g, b = x[0], x[1]
        # print(g,b)

        mu = df['value'].mean()
        std = df['value'].std()

        upper = df[ (df['value'] > (mu + up_limit * std)) ]
        upper = upper.label.groupby(upper['label']).count()
        lower = df[ (df['value'] < (mu - low_limit * std)) ]
        lower = lower.label.groupby(lower['label']).count()

#         print(fname)
    #     print(upper)
    #     print(lower)

        upper = format_df(upper)
        lower = format_df(lower)


#         summary = summary.append({'St': g, 'Anom': b, 'St_up': upper['freq'][0], 'Anom_up': upper['freq'][1], 'St_down': lower['freq'][0], 'Anom_down': lower['freq'][1]}, ignore_index=True)

        up_ratio = (upper['freq'][1] - upper['freq'][0]) / b
        up_prec  = upper['freq'][1]/(upper['freq'][1] + upper['freq'][0])
        low_ratio = (lower['freq'][1] - lower['freq'][0]) / b
        low_prec  = lower['freq'][1]/(lower['freq'][1] + lower['freq'][0])
        
        

        
        if up_ratio > arr[i][2]:
            arr[i][0] = fname
            arr[i][1] = up_limit
            arr[i][2] = up_ratio
            arr[i][3] = up_prec
            
        if low_ratio > arr[i][5]:
            arr[i][0] = fname
            arr[i][4] = low_limit
            arr[i][5] = low_ratio
            arr[i][6] = low_prec
            
        i += 1
        
#         summary.to_csv('Data' + str(up_limit) +'.csv')
        
#     df['Pred'] = 0
#     df.loc[ ( (df['value'] > (mu + 1.7 * std)) | (df['value'] < (mu - 2.1 * std)) ), 'Pred' ] = 1
    
# print(summary)
```
