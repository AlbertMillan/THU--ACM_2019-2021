```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
hidden_layers_out = [32, 64]
```

```python
train_loss_hist_0 = np.load("results/train_loss__" + str(hidden_layers_out[0]) + "_" + str(hidden_layers_out[1]) + "__dr1.npy")
test_loss_hist_0 = np.load("results/test_loss__" + str(hidden_layers_out[0]) + "_" + str(hidden_layers_out[1]) + "__dr1.npy")
train_acc_hist_0 = np.load("results/train_acc__" + str(hidden_layers_out[0]) + "_" + str(hidden_layers_out[1]) + "__dr1.npy")
test_acc_hist_0 = np.load("results/test_acc__" + str(hidden_layers_out[0]) + "_" + str(hidden_layers_out[1]) + "__dr1.npy")

train_loss_hist_a = np.load("results/train_loss__" + str(hidden_layers_out[0]) + "_" + str(hidden_layers_out[1]) + "__dr1_base.npy")
test_loss_hist_a = np.load("results/test_loss__" + str(hidden_layers_out[0]) + "_" + str(hidden_layers_out[1]) + "__dr1_base.npy")
train_acc_hist_a = np.load("results/train_acc__" + str(hidden_layers_out[0]) + "_" + str(hidden_layers_out[1]) + "__dr1_base.npy")
test_acc_hist_a = np.load("results/test_acc__" + str(hidden_layers_out[0]) + "_" + str(hidden_layers_out[1]) + "__dr1_base.npy")

train_loss_hist_b = np.load("results/train_loss__" + str(hidden_layers_out[0]) + "_" + str(hidden_layers_out[0]) + "__dr1.npy")
test_loss_hist_b = np.load("results/test_loss__" + str(hidden_layers_out[0]) + "_" + str(hidden_layers_out[0]) + "__dr1.npy")
train_acc_hist_b = np.load("results/train_acc__" + str(hidden_layers_out[0]) + "_" + str(hidden_layers_out[0]) + "__dr1.npy")
test_acc_hist_b = np.load("results/test_acc__" + str(hidden_layers_out[0]) + "_" + str(hidden_layers_out[0]) + "__dr1.npy")

train_loss_hist_c = np.load("results/train_loss__" + str(hidden_layers_out[1]) + "_" + str(hidden_layers_out[1]) + "__dr1.npy")
test_loss_hist_c = np.load("results/test_loss__" + str(hidden_layers_out[1]) + "_" + str(hidden_layers_out[1]) + "__dr1.npy")
train_acc_hist_c = np.load("results/train_acc__" + str(hidden_layers_out[1]) + "_" + str(hidden_layers_out[1]) + "__dr1.npy")
test_acc_hist_c = np.load("results/test_acc__" + str(hidden_layers_out[1]) + "_" + str(hidden_layers_out[1]) + "__dr1.npy")

# Drop-out
train_loss_hist_d = np.load("results/train_loss__" + str(hidden_layers_out[0]) + "_" + str(hidden_layers_out[1]) + "__dr0.npy")
test_loss_hist_d = np.load("results/test_loss__" + str(hidden_layers_out[0]) + "_" + str(hidden_layers_out[1]) + "__dr0.npy")
train_acc_hist_d = np.load("results/train_acc__" + str(hidden_layers_out[0]) + "_" + str(hidden_layers_out[1]) + "__dr0.npy")
test_acc_hist_d = np.load("results/test_acc__" + str(hidden_layers_out[0]) + "_" + str(hidden_layers_out[1]) + "__dr0.npy")
```

```python
plt.figure(figsize=(12,8))
plt.plot(np.arange(len(train_loss_hist_a)), train_loss_hist_a, label='(32,64)')
plt.plot(np.arange(len(train_loss_hist_b)), train_loss_hist_b, label='(32,32)')
plt.plot(np.arange(len(train_loss_hist_c)), train_loss_hist_c, label='(64,64)')
plt.plot(np.arange(len(train_loss_hist_d)), train_loss_hist_d, label='(32,64) Dropout 0.2')
# plt.plot(np.arange(len(train_loss_hist_0)), train_loss_hist_0, label='(32,64) No Momentum')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss/Epoch")
plt.legend()
plt.savefig("graphs/train_loss.png", bbox_inches="tight")
plt.show()

plt.figure(figsize=(12,8))
plt.plot(np.arange(len(train_acc_hist_a)), train_acc_hist_a, label='(32,64)')
plt.plot(np.arange(len(train_acc_hist_b)), train_acc_hist_b, label='(32,32)')
plt.plot(np.arange(len(train_acc_hist_c)), train_acc_hist_c, label='(64,64)')
plt.plot(np.arange(len(train_acc_hist_d)), train_acc_hist_d, label='(32,64) Dropout 0.2')
# plt.plot(np.arange(len(train_acc_hist_0)), train_acc_hist_0, label='(32,64) No Momentum')
plt.xlabel("Epochs")
plt.ylabel("Acc")
plt.ylim(None,1.0)
plt.title("Acc/Epoch")
plt.legend()
plt.savefig("graphs/train_acc.png", bbox_inches="tight")
plt.show()

print(">>> (32, 64):", np.max(train_acc_hist_a))
print(">>> (32, 32):", np.max(train_acc_hist_b))
print(">>> (64, 64):", np.max(train_acc_hist_c))
print(">>> (32, 64) Dropout:", np.max(train_acc_hist_d))


plt.figure(figsize=(12,8))
plt.plot(np.arange(len(test_loss_hist_a)), test_loss_hist_a, label='(32,64)')
plt.plot(np.arange(len(test_loss_hist_b)), test_loss_hist_b, label='(32,32)')
plt.plot(np.arange(len(test_loss_hist_c)), test_loss_hist_c, label='(64,64)')
plt.plot(np.arange(len(test_loss_hist_d)), test_loss_hist_d, label='(32,64) Dropout 0.2')
# plt.plot(np.arange(len(test_loss_hist_0)), test_loss_hist_0, label='(32,64) No Momentum')
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss/Epoch")
plt.legend()
plt.savefig("graphs/test_loss.png", bbox_inches="tight")
plt.show()

plt.figure(figsize=(12,8))
plt.plot(np.arange(len(test_acc_hist_a)), test_acc_hist_a, label='(32,64)')
plt.plot(np.arange(len(test_acc_hist_b)), test_acc_hist_b, label='(32,32)')
plt.plot(np.arange(len(test_acc_hist_c)), test_acc_hist_c, label='(64,64)')
plt.plot(np.arange(len(test_acc_hist_d)), test_acc_hist_d, label='(32,64) Dropout 0.2')
# plt.plot(np.arange(len(train_acc_hist_0)), train_acc_hist_0, label='(32,64) No Momentum')
plt.xlabel("Epochs")
plt.ylabel("Acc")
plt.ylim(None,1.0)
plt.title("Acc/Epoch")
plt.legend()
plt.savefig("graphs/test_acc.png", bbox_inches="tight")
plt.show()

print(">>> (32, 64):", np.max(test_acc_hist_a))
print(">>> (32, 32):", np.max(test_acc_hist_b))
print(">>> (64, 64):", np.max(test_acc_hist_c))
print(">>> (32, 64) Dropout:", np.max(test_acc_hist_d))
```
